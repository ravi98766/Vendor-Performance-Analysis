# Which vendors and brands demonstrate the highest sales performance?

```python
df.columns
```

```python
def format_dollars(value):
    if value >= 1_000_000:
        return f"{value / 1_000_000:.2f}M"
    elif value >= 1_000:
        return f"{value / 1_000:.2f}K"
    else:
        return str(value)
```

```python
top_vendors=df.groupby('VendorName')['TotalSalesDollars'].sum().nlargest(10)
top_brands=df.groupby('Description')['TotalSalesDollars'].sum().nlargest(10)
```

```python
top_vendors.apply(lambda x: format_dollars(x))
```

```python
top_brands.apply(lambda x: format_dollars(x))
```

```python
plt.figure(figsize=(15, 5))

# ---- Left plot ----
plt.subplot(1, 2, 1)
ax1 = sns.barplot(y=top_vendors.index, x=top_vendors.values, palette='Blues_r')

for bar in ax1.patches:
    ax1.text(
        bar.get_width() + (bar.get_width() * 0.02),
        bar.get_y() + bar.get_height() / 2,
        format_dollars(bar.get_width()),
        ha='left', va='center', color='black'
    )

plt.title('Top 10 Vendors by Sales', fontsize=18)

# ---- Right plot ----
plt.subplot(1, 2, 2)
ax2 = sns.barplot(y=top_brands.index, x=top_brands.values, palette='Reds_r')

for bar in ax2.patches:
    ax2.text(
        bar.get_width() + (bar.get_width() * 0.02),
        bar.get_y() + bar.get_height() / 2,
        format_dollars(bar.get_width()),
        ha='left', va='center', color='black'
    )

plt.title('Top 10 Brands by Sales', fontsize=18)

plt.tight_layout()
plt.show()
```

# Which vendors contribute the most to total purchase dollars?

```python
df.columns
```

```python
vendor_performance=df.groupby('VendorName').agg({'TotalPurchaseDollars':'sum','GrossProfit':'sum','TotalSalesDollars':'sum'}).reset_index()

vendor_performance['PercentageContribution%']=(vendor_performance['TotalPurchaseDollars']/vendor_performance['TotalPurchaseDollars'].sum())*100

# Rank Vendors by Total Purchase Dollars & Profitability
vendor_performance = round(vendor_performance.sort_values(by="TotalPurchaseDollars", ascending=False),2)
top_vendors=vendor_performance.head(10)
top_vendors['TotalPurchaseDollars']=vendor_performance['TotalPurchaseDollars'].apply(lambda x: format_dollars(x))
top_vendors['GrossProfit']=vendor_performance['GrossProfit'].apply(lambda x: format_dollars(x))
top_vendors['TotalSalesDollars']=vendor_performance['TotalSalesDollars'].apply(lambda x: format_dollars(x))
top_vendors
```

```python
top_vendors['CumulativeContribution%']=top_vendors['PercentageContribution%'].cumsum()
top_vendors
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

# Bar plot
sns.barplot(
    x=top_vendors['VendorName'],
    y=top_vendors['PercentageContribution%'],
    palette='Blues_r',
    ax=ax
)

# Adding text labels on bars
for i, value in enumerate(top_vendors['PercentageContribution%']):
    ax.text(
        i, value - 1, f"{value:.2f}%",  # formatted with 1 decimal place
        ha='center', va='center',
        fontsize=10, color='white'
    )

# Line plot for cumulative %
ax2 = ax.twinx()
ax2.plot(
    top_vendors['VendorName'],
    top_vendors['CumulativeContribution%'],
    marker='o', color='red', linestyle='--', label='Cumulative %'
)

# Legends and labels
ax.set_ylabel('Percentage Contribution %', fontsize=12)
ax2.set_ylabel('Cumulative Contribution %', fontsize=12)
ax.set_xlabel('Vendor Name', fontsize=12)
ax2.legend(loc='upper center')

ax.set_xticklabels(top_vendors['VendorName'], rotation=90)
plt.tight_layout()
plt.show()
```

# How much capital is locked in unsold inventory per vendor, and which vendors contribute the most to it?

```python
df.columns
```

```python
# Calculate Unsold Inventory Value
df['UnsoldInventoryValue']=(df['TotalPurchaseQuantity']-df['TotalSalesQuantity'])*df['PurchasePrice']
print("Total Unsold Capital:", format_dollars(df['UnsoldInventoryValue'].sum()))
```

```python
# Aggregate Capital Locked per Vendor
inventory_value_per_vendor = df.groupby('VendorName')[['UnsoldInventoryValue']].sum()

# Sort Vendors with the Highest Locked Capital
inventory_value_per_vendor = inventory_value_per_vendor.sort_values(by="UnsoldInventoryValue", ascending=False)
inventory_value_per_vendor['UnsoldInventoryValue'] = inventory_value_per_vendor['UnsoldInventoryValue'].apply(format_dollars)
inventory_value_per_vendor.head(10)
```
